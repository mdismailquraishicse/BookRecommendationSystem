```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
path_book= '/content/drive/MyDrive/MyProjects/ML Projects/Unsupervised ML/BookRecommendationSystem/data/Books.csv'
path_rating= '/content/drive/MyDrive/MyProjects/ML Projects/Unsupervised ML/BookRecommendationSystem/data/Ratings.csv'
path_users= '/content/drive/MyDrive/MyProjects/ML Projects/Unsupervised ML/BookRecommendationSystem/data/Users.csv'

book_data= pd.read_csv(path_book)
rating_data= pd.read_csv(path_rating)
users_data= pd.read_csv(path_users)
data_dict= {'book_data':book_data, 'rating_data':rating_data, 'users_data':users_data}
```
